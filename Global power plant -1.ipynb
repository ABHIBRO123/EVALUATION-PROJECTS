```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV files into DataFrames
data_usa = pd.read_csv('https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_USA.csv')
data_ind = pd.read_csv('https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_IND.csv')
data_aus = pd.read_csv('https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_AUS.csv')

# Concatenate the DataFrames
df = pd.concat([data_usa, data_ind, data_aus], axis=0)

# Check for and remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

# Optionally, display the first few rows of the concatenated DataFrame
print(df.head())

# Define the list of continuous columns
continous_columns = ['capacity_mw', 'latitude', 'longitude', 'wepp_id']  # Example, adjust based on your DataFrame
```

```python
# Create a figure for the distribution plots
plt.figure(figsize=(15,12))
plotnumber = 1

# Loop through each continuous column and create a subplot for its distribution plot
for i in continous_columns:
    plt.subplot(3, 3, plotnumber)
    sns.histplot(df[i], kde=True)  # Using sns.histplot instead of sns.distplot
    plt.title(f'Distribution of {i}')
    plotnumber += 1

# Display the plots
plt.tight_layout()
plt.show()
```
